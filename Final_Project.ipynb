# USING MULTIPLE LINEAR REGRESSION

```python
X=k.loc[:,['Item_Visibility','Item_Fat_Content','Item_MRP','Item_Outlet_Sales']]
X
```

```python
x=X.iloc[:,:-1].values
y=X.iloc[:,3].values
```

```python
print(pd.DataFrame(x))
print(pd.DataFrame(y))
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
labelencoder=LabelEncoder()
x[:,1]=labelencoder.fit_transform(x[:,1])
```

```python
print(pd.DataFrame(x))
```

```python
onehotencoder=OneHotEncoder(categorical_features=[1])
x=onehotencoder.fit_transform(x).toarray()
```

```python
print(pd.DataFrame(x))
```

```python
from sklearn.cross_validation import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.3,random_state=0)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

# we are using x_train nd y_train to plot best fit line.this is called linear regression then after we predict y_test corresponding to x_test...nd comparision between actual nd predicted value nd we saw that our machine learning algorithm is working...

```python
print(pd.DataFrame({'Actual':list(y_test),'Predicted':list(y_pred)}))
```

```python
from sklearn import metrics
print("Mean absolute Error:",metrics.mean_absolute_error(y_test,y_pred))
print("Mean Squared Error:",metrics.mean_squared_error(y_test,y_pred))
print("Root Mean absolute Error:",np.sqrt(metrics.mean_absolute_error(y_test,y_pred)))
```

```python
print(pd.DataFrame(x).columns)
coeff_df=pd.DataFrame(regressor.coef_,pd.DataFrame(x).columns,columns=['Coeffient'])
coeff_df
```
